##Понятие датасета

**Датасет** ($dataset$) — это структурированная коллекция данных, которая может содержать различные типы информации, такие как числа, строки или изображения, организованные в виде таблиц, массивов или файлов. Он широко используется в анализе данных, машинном обучении и статистике для обучения моделей и проведения экспериментов.

##Виды датасетов по источнику происхождения

**Реальные датасеты** — это данные, собранные из реальных источников: опросы, датчики, метрики на онлайн-платформах и т. д. Они отражают действительность и содержат разнообразные варианты данных, что делает их полезными для анализа, тестирования и валидации моделей. Однако они могут содержать шум, пропуски или ошибки.

**Синтетические датасеты** — наборы данных, которые создаются искусственно с использованием алгоритмов или имитационных моделей. Они могут быть разработаны для конкретных задач, чтобы имитировать возможные сценарии, которые могут не быть представлены в реальных данных. Это позволяет избежать проблем с конфиденциальностью, а также обеспечить контроль над параметрами данных. В силу синтетической природы они не всегда отражают всевозможные особенности реальных явлений.

##Продвинутый гуглинг

image.png

Современные **поисковые системы** поддерживают функционал для сложных поисковых запросов.

1. Использование **логических операторов**: `AND`, `OR`, `NOT` позволяют точно формулировать запросы и управлять содержимым результатов.

2. **Тип файлов**: `filetype:csv`, `filetype:xlsx` и т. д. помогают ограничить поиск определенными форматами данных.

3. **Данные о времени**: `after:2025-01-01` позволяет находить данные, опубликованные после указанной даты.

4. **Исключение лишнего**: оператор `-` позволяет исключать сайты с отдельными словами из выдачи. Для исключения словоформ можно использовать запись `-слов*`

Собрав все части вместе, получаем продвинутый запрос для поиска датасетов:

```
(примерное описание поиска) (filetype:csv OR filetype:xlsx OR filetype:json OR filetype:txt OR filetype:zip) after:2025-01-01 -ключевое_слово*
```

##Библиотека Pandas

image.png

`Pandas` — библиотека Python для обработки и анализа табличных данных, изначально разработанная Уэсом МакКинни (Wes McKinney) в 2008 году и выпущенная как открытый проект.

`Pandas` построена поверх NumPy и предоставляет удобные структуры данных (`Series`, `DataFrame`) и операции для манипуляций с таблицами и временными рядами. Обеспечивает удобное чтение/запись CSV/Excel/JSON, мощные методы группировки и агрегации, работу с пропусками, объединение таблиц и преобразования типов. Может использоваться для веб-скрейпинга.

##Понятие CSV

image.png

`CSV` (Comma-Separated Values) — это простой текстовый формат для представления **табличных данных**. В файлах CSV строки представляют собой записи, а столбцы разделены запятыми или другим символом-сепаратором.

Этот формат популярен благодаря своей простоте и совместимости, но не поддерживает сложные структуры данных или типы, что ограничивает его использование в более сложных задачах.

##Скрейпинг и обработка CSV с использованием Pandas. Сбор данных о планетах Солнечной системы

image.png

###Постановка задачи

На [сайте](http://www.allplanets.ru/solar_sistem.htm) приведены основные данные о планетах Солнечной системы. Необходимо скачать эти табличные данные, сохранить в формате `CSV` и проверить корректность сохранения.

###Решение

Испортируем `pandas`. Скачаем таблицу по ссылке. Аккуратно обработаем кодировку.

````python
import pandas as pd

url = "http://www.allplanets.ru/solar_sistem.htm"
df = pd.read_html(url, encoding='windows-1251')[0].T
````

Рассмотрим полученные данные.

````python
df.head()
````

От номеров колонок перейдём к их названиям. А строки в таблице будем индексировать по названию планет.

````python
df.columns = df.iloc[0].values
df = df.drop(index=0)
df.set_index('Планета')
df.head()
````

Сохраним таблицу в `CSV`.

````python
df.to_csv('planets.csv', sep=";", index=False)
````

Считаем таблицу в оперативную память.

````python
df_from_csv = pd.read_csv('planets.csv', sep=";")
df_from_csv.head()
````

###Постановка задачи

[Датасет](https://www.kaggle.com/datasets/dylanjcastillo/7k-books-with-metadata) с Kaggle содержит данные о **7 тысячах книг**. Этот сборник представляет собой модификацию классического датасета GoogleReads, дополненного актуальными данными, найденными с использованием ISBN.

Приведены идентификаторы ISBN (как 10, так и 13), заголовок, подзаголовок, авторы, категории, URL, описание, год издания и средний рейтинг на Goodreads

Для выполнения задания необходимо:

1. Скачать датасет.
2. Прочитать CSV.
3. Разделить книги по категориям.
4. Сохранить каждую часть в отдельный CSV.
5. Считать части по отдельности.
6. Объединить обратно в один CSV.
7. Проверить корректность преобразований.

###Решение задачи

Скачаем датасет с использованием библиотеки `opendatasets`.

````python
!pip install -q opendatasets
````

Сохраним токен Kaggle в необходимой для opendatasets директории.

````python
from google.colab import files
import os

uploaded = files.upload()
````

Скачаем датасет.

````python
import opendatasets as od

url = "https://www.kaggle.com/datasets/dylanjcastillo/7k-books-with-metadata"
od.download(url)
````

Чтение `CSV` с pandas. Файл считается в оперативную память и сформирует таблицу.

````python
import pandas as pd

df = pd.read_csv("7k-books-with-metadata/books.csv")
````

````python
df.head()
````

Выведем список уникальных категорий.

````python
df["categories"].nunique()
````

Подсчитаем книги с неуказанной категорией.

````python
df["categories"].isna().sum()
````

Отнесём их к `Unknown`.

````python
df["categories"] = df["categories"].fillna("Unknown")
````

Создадим папку `books_by_category`, сохраним книги каждой категории в отдельные файлы `CSV`.

````python
import os

os.makedirs("books_by_category", exist_ok=True)

unique_cats = df["categories"].unique()
for cat in unique_cats:
    safe_name = str(cat).replace(" ", "_")
    part = df[df["categories"] == cat]
    part.to_csv(f"books_by_category/{safe_name}.csv", index=False)
````

Проверим, выполнился ли код.

````python
os.listdir('./books_by_category')
````

Перейдём к следующей части задания — работа с несколькими `CSV`. Получим список путей ко всем созданным в папке `books_by_category` `CSV`-файлам.

````python
import glob

files = glob.glob("books_by_category/*.csv")
````

Считаем и объединим обратно в один `DataFrame`.

````python
parts = [pd.read_csv(f) for f in files]
df_combined = pd.concat(parts, ignore_index=True)
df_combined.head()
````

Проверка соответствия исходного и собранного `DataFrame` по количеству строк:

````python
assert df.shape[0] == df_combined.shape[0]
````

Успешно. Размерности совпали.

##Пример работы с открытыми API.  Сбор данных о научных статьях по искусственному интеллекту в физике

images.png

###Постановка задачи

Необходимо получить данные о научных статьях, посвящённых применению искусственного интеллекта в физике, из архива **Semantic Scholar** за 2025 год. Полученные данные следует преобразовать в `DataFrame` и сохранить в `CSV`.

###Решение задачи

Укажем ссылку, по которой можно взаимодействовать с открытым API Semantic Scholar. Опишем запрос. Правила написания параметров запроса (ключи и значения) следует смотреть на сайте разработчиков API.

````python
url = "http://api.semanticscholar.org/graph/v1/paper/search/bulk"

query_params = {
    "query": 'ai, physics',
    "fields": "title,url,publicationTypes,publicationDate,openAccessPdf",
    "year": "2025-",
    "limit": "1000"
}
````


Сделаем GET-запрос к API.

````python
import requests

response = requests.get(url, params=query_params).json()['data']
````

Преобразуем результат в `DataFrame`.

````python
df = pd.DataFrame(response)
df.head()
````

Сохраним в `CSV`.

````python
df.to_csv('1k-ai-physics-articles-2025.csv', index=False)
````
